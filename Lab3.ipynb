```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# 1. Загрузка и быстрая очистка (из ЛР1)
url = "https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv"
df = pd.read_csv(url)
df.fillna(df.median(numeric_only=True), inplace=True)
df.fillna('Unknown', inplace=True)

# 2. Кодирование и масштабирование (из ЛР2)
le = LabelEncoder()
df['Cylinders'] = le.fit_transform(df['Cylinders'].astype(str))
df = pd.get_dummies(df, columns=['Type'], prefix='Type')

# Создаем новые признаки и масштабируем
df['HP_per_EngineSize'] = df['Horsepower'] / df['EngineSize']
scaler = MinMaxScaler()
cols = ['Price', 'Horsepower', 'Weight', 'EngineSize', 'HP_per_EngineSize']
df[cols] = scaler.fit_transform(df[cols])

# 3. Формируем тот самый df_final
df_final = df.drop(columns=['Model', 'Manufacturer', 'Make'])

print("Готово! Теперь переменная df_final определена и можно обучать модель.")
```
