```python
# 1. Импорт библиотек (это база для любой лабы)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Загрузка данных
# Если вы еще не скачали файл, можно загрузить тестовый датасет прямо из интернета
url = "https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv"
df = pd.read_csv(url)

# 3. Посмотрим, что внутри
print("Размер таблицы:", df.shape)
df.head()
```

Лабораторная работа №1
Тема: Исследовательский анализ данных (EDA)
Цель: Анализ характеристик автомобилей и подготовка данных.

```python
# Посчитаем количество пропусков в каждой колонке
missing_values = df.isnull().sum()
print("Пропуски по столбцам:")
print(missing_values[missing_values > 0])

# Визуализация пропусков (1-й метод по ТЗ)
plt.figure(figsize=(12, 4))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title("Карта пропущенных значений")
plt.show()
```

```python
def clean_data(data):
    df_copy = data.copy()

    # 2-й метод обработки пропусков:
    # Числовые признаки заполняем медианой, чтобы не искажать статистику
    numeric_cols = df_copy.select_dtypes(include=[np.number]).columns
    df_copy[numeric_cols] = df_copy[numeric_cols].fillna(df_copy[numeric_cols].median())

    # Категориальные (текстовые) заполняем значением 'Unknown'
    object_cols = df_copy.select_dtypes(include=['object']).columns
    df_copy[object_cols] = df_copy[object_cols].fillna('Unknown')

    return df_copy

# Применяем функцию
df_cleaned = clean_data(df)
print("Пропусков после очистки:", df_cleaned.isnull().sum().sum())
```

```python
# Посмотрим на выбросы в колонке Price
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_cleaned['Price'])
plt.title("Анализ выбросов для цены (Price)")
plt.show()
```

В колонке Price обнаружены выбросы, машины дороже 40-50 тысяч встречаются редко

```python
# 1. Признак: Разница между максимальной и минимальной ценой (ценовой диапазон модели)
df_cleaned['Price_Range'] = df_cleaned['Max.Price'] - df_cleaned['Min.Price']

# 2. Признак: Средний расход топлива (среднее между городом и трассой)
df_cleaned['Avg_MPG'] = (df_cleaned['MPG.city'] + df_cleaned['MPG.highway']) / 2

# 3. Признак: Удельная мощность (мощность на единицу веса)
# Weight в фунтах, Horsepower в л.с.
df_cleaned['HP_per_Weight'] = df_cleaned['Horsepower'] / df_cleaned['Weight']

print("Новые признаки добавлены. Теперь столбцов:", df_cleaned.shape[1])
df_cleaned[['Price_Range', 'Avg_MPG', 'HP_per_Weight']].head()
```

```python
# Выбираем только числовые колонки для корреляции
numeric_df = df_cleaned.select_dtypes(include=[np.number])

plt.figure(figsize=(12, 10))
sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f", cmap='coolwarm', center=0)
plt.title("Матрица корреляции признаков")
plt.show()
```

Цена автомобиля имеет прямую зависимость от мощности двигателя (Horsepower)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cleaned, x='Horsepower', y='Price', hue='Type')
plt.title("Зависимость цены от мощности двигателя")
plt.grid(True)
plt.show()
```
